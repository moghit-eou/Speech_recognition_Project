```python
import sounddevice as sd
import numpy as np
import librosa
import tensorflow as tf
import os
import scipy.io.wavfile as wav
import numpy as np
import librosa, librosa.display
import matplotlib.pyplot as plt

duration = 2  # Durée de l'enregistrement (en secondes)
sr = 22050  # Taux d'échantillonnage
n_fft = 2048
hop_length = 512
fixed_length = 10000  # Taille des vecteurs 1D
def record_audio(duration, sample_rate = sr):
    duration = 1
    print("Recording...")
    audio = sd.rec(int(duration * sample_rate), samplerate=sample_rate, channels=1, dtype=np.float32)
    sd.wait()
    print("Recording finished.")
    audio_path = "source/output1.wav"
    wav.write(audio_path, sample_rate, audio)
    return audio
```

```python
audio = record_audio(2 )
```

```python
max(audio)
```

```python
plt.figure(figsize=(15,3))
librosa.display.waveshow(audio.flatten(), sr=1, alpha=0.5)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Waveform")
```
